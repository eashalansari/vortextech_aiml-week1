## Objective
This notebook demonstrates data cleaning and exploratory data analysis using the Titanic dataset.

## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Titanic dataset
df = pd.read_csv("titanic.csv")

# Display the first 5 rows
df.head()
```

## Load Dataset

```python
# Check dataset information
df.info()
```

# Explore Dataset

```python
# Display the shape of the dataset
df.shape
```

# Check missing values

```python
# Check missing values
df.isnull().sum()
```

# Handle missing values

```python
# Fill missing values in Age with the median
df["Age"] = df["Age"].fillna(df["Age"].median())

# Fill missing value in Fare with the median
df["Fare"] = df["Fare"].fillna(df["Fare"].median())

# Drop Cabin column because it has too many missing values
df.drop("Cabin", axis=1, inplace=True)
```

# Explanation

Age column had 86 missing values, so I filled them with the median because it is less affected by outliers.

Fare had only 1 missing value, so I filled it with the median.

Cabin had 327 missing values, so I dropped the column because most of its data was missing and it was not useful for analysis.

# Remove the missing the values

# Count passengers

```python
# Count of male and female passengers
df["Sex"].value_counts()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.hist(df["Age"], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
import matplotlib.pyplot as plt

df["Sex"].value_counts().plot(kind="bar")

plt.title("Number of Male and Female Passengers")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```
